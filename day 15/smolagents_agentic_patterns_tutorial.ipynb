## Pattern A: Prompt Chaining

### Concept
**Prompt Chaining** connects multiple agents in a sequential workflow where the output of one agent becomes the input to the next. This pattern is ideal for:
- Multi-step transformations
- Progressive refinement
- Staged processing pipelines

### Example: Text Summarization â†’ Translation Pipeline
We'll create a chain where:
1. **Agent A** (Summarizer) condenses a long text
2. **Agent B** (Translator) translates the summary to another language

```python
from smolagents import CodeAgent, ToolCallingAgent

# Create Agent A: Text Summarizer (using ToolCallingAgent for simpler text generation)
agent_summarizer = ToolCallingAgent(
    tools=[],
    model=model,
    max_steps=2
)

# Create Agent B: Translator (using ToolCallingAgent for text generation)
agent_translator = ToolCallingAgent(
    tools=[],
    model=model,
    max_steps=2
)

# Sample long text to process
long_text = """
Artificial Intelligence has revolutionized numerous industries over the past decade.
From healthcare diagnostics to autonomous vehicles, AI systems are becoming increasingly
sophisticated. Machine learning algorithms can now detect diseases from medical images
with accuracy rivaling human experts. Natural language processing has enabled chatbots
and virtual assistants that can understand and respond to human queries. Computer vision
systems power facial recognition and object detection in security systems. The field
continues to evolve rapidly with new breakthroughs in deep learning, reinforcement
learning, and neural architecture search.
"""

print("=" * 60)
print("PATTERN A: PROMPT CHAINING")
print("=" * 60)

# Step 1: Agent A summarizes the text
print("\nğŸ”— Step 1: Summarizing text...")
summary = agent_summarizer.run(
    f"Summarize this text in 2 sentences:\n\n{long_text}"
)
print(f"\nğŸ“ Summary:\n{summary}")

# Step 2: Agent B translates the summary
print("\nğŸ”— Step 2: Translating summary to Spanish...")
translation = agent_translator.run(
    f"Translate the following English text to Spanish. Return only the Spanish translation, nothing else:\n\n{summary}"
)
print(f"\nğŸŒ Translation:\n{translation}")

print("\nâœ… Prompt Chain Complete!")
```

## Pattern B: Routing

### Concept
**Routing** uses an intelligent dispatcher that analyzes incoming requests and directs them to the appropriate specialized tool or agent. This pattern enables:
- Efficient task delegation
- Specialized processing based on request type
- Modular, maintainable architectures

### Example: Query Router with Math vs. General Chat Tools
We'll create a router that decides whether to use a Math Tool or a General Chat Tool based on the query content.

```python
from smolagents import tool

# Define specialized tools using the @tool decorator

@tool
def math_calculator(expression: str) -> str:
    """
    Evaluates mathematical expressions safely.

    Args:
        expression: A mathematical expression as a string (e.g., "2 + 2", "10 * 5")

    Returns:
        The result of the calculation as a string
    """
    try:
        # Safe evaluation of mathematical expressions
        result = eval(expression, {"__builtins__": {}}, {})
        return f"Result: {result}"
    except Exception as e:
        return f"Error calculating: {str(e)}"


@tool
def general_chat(query: str) -> str:
    """
    Handles general conversation and non-mathematical queries.

    Args:
        query: The user's general question or statement

    Returns:
        A friendly conversational response
    """
    return f"I received your query: '{query}'. I'm a general chat assistant ready to help with information, advice, or conversation!"


# Create Router Agent with both specialized tools
router_agent = CodeAgent(
    tools=[math_calculator, general_chat],
    model=model,
    max_steps=3
)

print("=" * 60)
print("PATTERN B: ROUTING")
print("=" * 60)

# Test queries of different types
test_queries = [
    "What is 125 multiplied by 8?",
    "Tell me about the weather today",
    "Calculate 2 ** 10",
    "What's your favorite color?"
]

for query in test_queries:
    print(f"\nğŸ“¥ Query: {query}")

    # The router agent will automatically select the appropriate tool
    result = router_agent.run(query)

    print(f"ğŸ“¤ Response: {result}")
    print("-" * 60)

print("\nâœ… Routing Complete!")
```

## Pattern C: Parallelization

### Concept
**Parallelization** executes multiple agents or tasks concurrently to improve efficiency. This pattern is valuable for:
- Independent tasks that can run simultaneously
- Reducing total execution time
- Gathering diverse perspectives on the same input

### Example: Pros & Cons Analysis
We'll run two agents in parallelâ€”one generating pros and another generating cons for a given topic, then aggregate their results.

```python
import concurrent.futures
from typing import Dict

# Create specialized agents for pros and cons analysis
# Using ToolCallingAgent for cleaner text generation
agent_pros = ToolCallingAgent(tools=[], model=model, max_steps=2)
agent_cons = ToolCallingAgent(tools=[], model=model, max_steps=2)

def generate_pros(topic: str) -> str:
    """Generate advantages/pros for a given topic"""
    prompt = f"List 3 key advantages or pros of {topic}. Be concise and specific. Provide a numbered list with brief explanations."
    return agent_pros.run(prompt)

def generate_cons(topic: str) -> str:
    """Generate disadvantages/cons for a given topic"""
    prompt = f"List 3 key disadvantages or cons of {topic}. Be concise and specific. Provide a numbered list with brief explanations."
    return agent_cons.run(prompt)

def parallel_analysis(topic: str) -> Dict[str, str]:
    """
    Run pros and cons analysis in parallel using ThreadPoolExecutor
    """
    print(f"ğŸ”„ Starting parallel analysis for: {topic}")

    # Execute both agents concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=2) as executor:
        # Submit both tasks
        future_pros = executor.submit(generate_pros, topic)
        future_cons = executor.submit(generate_cons, topic)

        # Wait for both to complete and gather results
        pros = future_pros.result()
        cons = future_cons.result()

    return {"pros": pros, "cons": cons}


print("=" * 60)
print("PATTERN C: PARALLELIZATION")
print("=" * 60)

# Test topic
topic = "Remote Work"

# Run parallel analysis
results = parallel_analysis(topic)

# Display aggregated results
print(f"\nğŸ“Š Analysis Results for '{topic}':")
print("\nâœ… PROS:")
print(results["pros"])
print("\nâŒ CONS:")
print(results["cons"])

print("\nâœ… Parallel Analysis Complete!")
```

## Pattern D: Orchestrator-Worker

### Concept
The **Orchestrator-Worker** pattern uses a manager agent that decomposes complex tasks into subtasks and delegates them to worker agents. This hierarchical structure enables:
- Complex task decomposition
- Coordinated multi-step workflows
- Clear separation of planning and execution

### Example: Blog Post Creation Pipeline
A Manager Agent breaks down "Write a blog post" into subtasks (outline, introduction, body, conclusion) and assigns each to a Worker Agent.

```python
from typing import List
import re
from openai import OpenAI

def orchestrator_worker_pattern(main_task: str) -> str:
    """
    Implements the Orchestrator-Worker pattern:
    1. Manager decomposes the task into subtasks
    2. Worker executes each subtask step-by-step
    3. Results are aggregated into final output

    Note: This pattern uses direct OpenAI API calls for cleaner output without agent verbosity
    """

    print(f"ğŸ¯ Main Task: {main_task}\n")

    # Initialize OpenAI client
    client = OpenAI(
        api_key=os.environ["OPENAI_API_KEY"],
        base_url=os.environ.get("OPENAI_BASE_URL")
    )

    # Step 1: Manager decomposes the task
    print("ğŸ“‹ Manager: Decomposing task into subtasks...")
    decomposition_prompt = f"""Break down this task into 3 clear, actionable subtasks: "{main_task}"
Return ONLY a numbered list (1., 2., 3.) with no additional text."""

    manager_response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": decomposition_prompt}]
    )

    subtasks_text = manager_response.choices[0].message.content
    print(f"âœ“ Subtasks identified:\n{subtasks_text}\n")

    # Parse subtasks into list
    subtasks = [line.strip() for line in subtasks_text.split('\n')
                if line.strip() and re.match(r'^\d+\.', line.strip())]

    # Step 2: Worker executes each subtask sequentially
    print("ğŸ‘· Worker: Executing subtasks...\n")
    results = []
    conversation_history = []

    for i, subtask in enumerate(subtasks, 1):
        print(f"ğŸ”¨ Subtask {i}: {subtask.split('.', 1)[1].strip() if '.' in subtask else subtask}")

        # Build context-aware prompt
        if results:
            context = "\n\n".join([f"Step {j+1} output: {results[j]}" for j in range(len(results))])
            prompt = f"""Main task: {main_task}

Previous work:
{context}

Now complete: {subtask}

Build upon previous work. Provide only the output for this step."""
        else:
            prompt = f"Main task: {main_task}\n\nComplete: {subtask}\n\nProvide only the output for this step."

        # Execute subtask
        conversation_history.append({"role": "user", "content": prompt})
        worker_response = client.chat.completions.create(
            model="gpt-4.1-mini",
            messages=conversation_history
        )

        result = worker_response.choices[0].message.content
        conversation_history.append({"role": "assistant", "content": result})
        results.append(result)

        print(f"âœ“ Completed\n")

    # Step 3: Manager selects the final result (last subtask is typically the polished version)
    print("ğŸ“¦ Manager: Selecting final deliverable...\n")

    # For blog post writing, the last step is usually the final polished version
    # Return only the last result (the refined, publication-ready output)
    final_output = results[-1]

    return final_output


print("=" * 60)
print("PATTERN D: ORCHESTRATOR-WORKER")
print("=" * 60)
print()

# Complex task to decompose
task = "Write a short blog post about the benefits of exercise"

# Execute orchestrator-worker pattern
final_result = orchestrator_worker_pattern(task)

print("=" * 60)
print("ğŸ“„ FINAL BLOG POST:")
print("=" * 60)
print(final_result)
print("\nâœ… Orchestrator-Worker Pattern Complete!")
```

## Pattern E: Evaluator-Optimizer

### Concept
The **Evaluator-Optimizer** pattern implements a feedback loop where:
1. A Generator Agent creates an initial solution
2. An Evaluator Agent critiques the solution
3. The Generator improves based on feedback
4. This cycle repeats until quality criteria are met

This pattern enables:
- Iterative refinement
- Quality assurance through feedback
- Self-improving systems

### Example: Story Writing with Iterative Improvement
We'll create a story, have it evaluated, and then improve it based on the critique.

```python
# Create Generator and Evaluator agents
agent_generator = CodeAgent(tools=[], model=model, max_steps=3)
agent_evaluator = CodeAgent(tools=[], model=model, max_steps=3)

def evaluator_optimizer_loop(task: str, max_iterations: int = 2) -> Dict[str, any]:
    """
    Implements the Evaluator-Optimizer pattern with feedback loop:
    1. Generator creates initial solution
    2. Evaluator provides critique
    3. Generator improves based on feedback
    4. Repeat for max_iterations
    """

    print(f"ğŸ¯ Task: {task}\n")

    # Initial generation
    print("ğŸ¨ Iteration 1: Initial Generation")
    print("-" * 60)

    current_solution = agent_generator.run(task)
    print(f"ğŸ“ Generated:\n{current_solution}\n")

    # Iterative improvement loop
    for iteration in range(max_iterations):
        print(f"\nğŸ” Iteration {iteration + 2}: Evaluation & Optimization")
        print("-" * 60)

        # Evaluator critiques the current solution
        evaluation_prompt = f"""
        Evaluate this response to the task "{task}":

        {current_solution}

        Provide specific, constructive feedback on how to improve it.
        Focus on clarity, engagement, and completeness.
        """

        critique = agent_evaluator.run(evaluation_prompt)
        print(f"ğŸ’¬ Critique:\n{critique}\n")

        # Generator improves based on critique
        improvement_prompt = f"""
        Original task: {task}

        Previous version:
        {current_solution}

        Feedback received:
        {critique}

        Rewrite and improve the response based on this feedback.
        """

        current_solution = agent_generator.run(improvement_prompt)
        print(f"âœ¨ Improved Version:\n{current_solution}\n")

    return {
        "final_solution": current_solution,
        "iterations": max_iterations + 1
    }


print("=" * 60)
print("PATTERN E: EVALUATOR-OPTIMIZER")
print("=" * 60)
print()

# Task for iterative improvement
writing_task = "Write a compelling 3-sentence story about a robot learning to paint"

# Execute evaluator-optimizer loop
result = evaluator_optimizer_loop(writing_task, max_iterations=2)

print("=" * 60)
print("ğŸ† FINAL RESULT:")
print("=" * 60)
print(result["final_solution"])
print(f"\nâœ… Completed after {result['iterations']} iterations!")
print("\nâœ… Evaluator-Optimizer Pattern Complete!")
```

## Summary of Agentic Patterns

Congratulations! ğŸ‰ You've explored 5 powerful agentic patterns with smolagents. Here's a quick reference guide:

| Pattern | When to Use | Key Benefits |
|---------|-------------|--------------|
| **Prompt Chaining** | Sequential transformations, multi-stage processing | Simple, predictable flow; easy to debug |
| **Routing** | Tasks requiring different expertise/tools | Efficient delegation; modular architecture |
| **Parallelization** | Independent tasks, diverse perspectives | Reduced latency; concurrent processing |
| **Orchestrator-Worker** | Complex tasks needing decomposition | Clear structure; manageable subtasks |
| **Evaluator-Optimizer** | Quality-critical outputs, iterative refinement | Self-improvement; quality assurance |

## Choosing the Right Pattern

- **Use Prompt Chaining** when you have a clear pipeline with dependent steps
- **Use Routing** when different request types need specialized handling  
- **Use Parallelization** when tasks are independent and time is critical
- **Use Orchestrator-Worker** when complexity requires hierarchical planning
- **Use Evaluator-Optimizer** when quality improvement through iteration is valuable
