```python
# place your code here

pc = Pinecone(api_key="pcsk_oqq5v_JsMNBGygcPN1ZNN7VHN1ofFSH9v8AaF1YYXc7tBR3cxjN4aHHGJCvLxyn3mUwEm")
model = SentenceTransformer('all-MiniLM-L6-v2')

index_name = "test"
if not pc.has_index(index_name):
    pc.create_index_for_model(
        name=index_name,
        cloud="aws",
        region="us-east-1",
        embed={
            "model":"llama-text-embed-v2",
            "field_map":{"text": "chunk_text"}
        }
    )

index = pc.Index(index_name)


texts = [
    "I love programming Python",
    "Machine learning is cool",
    "Math and Programming skills are important in computer science",
    "I like solving algorithmic problems"
]

vectors = []
for i, text in enumerate(texts):
    vector_values = model.encode(text).tolist()
    vectors.append((str(i), vector_values, {"text": text}))

index.upsert(vectors=vectors)
print("Data  uploaded!")

query_text = "Python"
query_vector = model.encode(query_text).tolist()

results = index.query(
    vector=query_vector,
    top_k=2,
    include_metadata=True
)

```

Print search results.

```python
# place your code here
for match in results['matches']:
    print(f"Score: {match['score']:.4f} | Text: {match['metadata']['text']}")
```
