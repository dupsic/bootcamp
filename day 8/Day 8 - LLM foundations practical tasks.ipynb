#### 16. Generate a Python code with API call.

Write a prompt that generates a Python function to calculate the factorial of a number. Once you have written the prompt, use it to call the API and generate the code. Test the generated code to ensure it calculates the factorial correctly for various inputs.

````python
prompt_16 = "Write a Python function called 'calculate_factorial' that takes an integer 'n' as an input. The function should return the factorial of n."

response_16 = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt_16}],
    temperature=0.1
)

print(response_16.choices[0].message.content)
````

````python
def calculate_factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    factorial = 1
    for i in range(2, n + 1):
        factorial *= i
    return factorial

print(calculate_factorial(2))
print(calculate_factorial(3))
print(calculate_factorial(5))
````
